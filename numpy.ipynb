````python
import numpy as np

# 創建一個形狀為 (3, 4) 的數組
a = np.arange(2, 14).reshape((3, 4))
print(a)
print("\n")

# 返回數組中最小值的索引（扁平化的索引）
min_index = np.argmin(a)
print(min_index)
print("\n")

# 返回數組中最大值的索引（扁平化的索引）
max_index = np.argmax(a)
print(max_index)
print("\n")

# 計算數組 a 的平均值
mean_value = np.mean(a)
print(mean_value)
print("\n")

# 也可以使用數組對象的 mean 方法計算平均值
mean_value_alt = a.mean()
print(mean_value_alt)
print("\n")

# 使用 np.average 函數計算加權平均值（在本例中未指定權重，因此等同於平均值）
weighted_mean = np.average(a)
print(weighted_mean)
print("\n")

# 計算數組 a 的中位數
median_value = np.median(a)
print(median_value)
print("\n")

# 複製並輸出原始數組 a
print(a)
print("\n")

# 計算數組元素的累積和，然後重新塑形為 (3, 4)
cumulative_sum = np.cumsum(a).reshape((3, 4))
print(cumulative_sum)
print("\n")

# 計算相鄰元素的差值
difference = np.diff(a)
print(difference)
print("\n")

# 創建一個降序的數組 b，形狀為 (3, 4)
b = np.arange(14, 2, -1).reshape((3, 4))
print(b)
print("\n")

# 對數組 b 進行排序，按行進行升序排序
sorted_b = np.sort(b)
print(sorted_b)
print("\n")

# 使用 transpose 函數進行轉置（矩陣的行和列互換）
transposed_b = np.transpose(b)
print(transposed_b)
print("\n")

# 也可以使用數組對象的 T 屬性進行轉置
transposed_b_alt = b.T
print(transposed_b_alt)
print("\n")

# 矩陣相乘：b 的轉置乘以 b
dot_product = transposed_b.dot(b)
print(dot_product)
print("\n")

# 將數組 b 中小於 4 的元素設為 4，大於 10 的元素設為 10
clipped_b = np.clip(b, 4, 10)
print(clipped_b)
print("\n")

# 計算數組 b 沿著軸 1（列）的平均值
mean_along_axis1 = np.mean(b, axis=1)
print(mean_along_axis1)
````

````python
import numpy as np

# 創建一維數組 a，範圍為 3 到 15（不包括 15）
a = np.arange(3, 15)
print("一維數組 a:")
print(a)

# 輸出索引為 3 的元素
print("索引為 3 的元素:", a[3])
print("\n")

# 創建二維數組 a，形狀為 (3, 4)，範圍為 3 到 15（不包括 15）
a = np.arange(3, 15).reshape(3, 4)
print("二維數組 a:")
print(a)

# 輸出索引為 2 的行
print("索引為 2 的行:", a[2])

# 輸出索引為 (1, 1) 的元素
print("索引為 (1, 1) 的元素:", a[1][1])
# 或者使用逗號分隔的索引
print("索引為 (1, 1) 的元素（另一種方法）:", a[1, 1])

# 輸出索引為 1 的整行
print("索引為 1 的整行:", a[1, :])

# 輸出索引為 1 的整列
print("索引為 1 的整列:", a[:, 1])

# 輸出索引為 1 的行中索引為 1 到 2 的元素
print("索引為 1 的行中索引為 1 到 2 的元素:", a[1, 1:3])
print("\n")

# 逐行迭代數組 a
print("逐行迭代:")
for row in a:
    print(row)
print("\n")

# 逐列迭代數組 a 的轉置
print("逐列迭代（轉置）:")
for column in a.T:
    print(column)
print("\n")

# 展平數組 a（轉為一維數組）
print("展平數組 a:")
print(a.flatten())
print("\n")

# 使用 flat 迭代數組 a 的每個元素
print("使用 flat 迭代元素:")
for element in a.flat:
    print(element)
````

````python
import numpy as np

# 創建兩個一維數組 a 和 b
a = np.array([1, 1, 1])
b = np.array([2, 2, 2])

# 使用 vstack 將兩個數組垂直堆疊
vstack_result = np.vstack((a, b))
print("垂直堆疊數組 a 和 b:")
print(vstack_result)
print("\n")

# 將垂直堆疊的結果賦值給新的變數 c，並輸出數組形狀
c = vstack_result
print("數組形狀 - a:", a.shape, "b:", b.shape, "c:", c.shape)
print("\n")

# 使用 hstack 將兩個數組水平堆疊
hstack_result = np.hstack((a, b))
print("水平堆疊數組 a 和 b:")
print(hstack_result)
print("數組形狀 - a:", a.shape, "b:", b.shape, "c:", c.shape, "d:", hstack_result.shape)
print("\n")

# 將一維數組 a 轉為二維數組，形狀為 (3, 1)
a_2d = a[:, np.newaxis]
print("重塑 a（添加新軸）:")
print(a_2d)
print("\n")

# 將二維數組 a_2d 轉置（沿著新的軸）
a_transposed = a[np.newaxis, :]
print("轉置 a_2d（使用新軸）:")
print(a_transposed)
print("轉置後的 a_2d 形狀:", a_transposed.shape)
print("\n")

# 創建兩個二維數組 a 和 b，形狀均為 (3, 1)
a = np.array([1, 1, 1])[:, np.newaxis]
b = np.array([2, 2, 2])[:, np.newaxis]

# 使用 concatenate 在現有軸上連接兩個數組
concatenated_result = np.concatenate((a, b, b, a))
print("沿軸 0 連接的數組:")
print(concatenated_result)

# 使用 concatenate 在新的軸上連接兩個數組
concatenated_result_axis1 = np.concatenate((a, b, b, a), axis=1)
print("\n沿軸 1 連接的數組:")
print(concatenated_result_axis1)
````

````python
import pandas as pd
import numpy as np

# 生成日期範圍
dates = pd.date_range('20230101', periods=6)

# 創建 DataFrame，行索引為日期，列索引為 'A', 'B', 'C', 'D'
df = pd.DataFrame(np.arange(24).reshape((6, 4)), index=dates, columns=['A', 'B', 'C', 'D'])
print("DataFrame:")
print(df)
print("\n")

# 選擇 'A' 列的兩種方法
print("選擇 'A' 列:")
print(df['A'])
print("\n")
print(df.A)
print("\n")

# 選擇前三行
print("選擇前三行:")
print(df[0:3])
print("\n")

# 用行索引選擇特定行
print("用行索引選擇 '20230102' 行:")
print(df.loc['20230102'])
print("\n")

# 用行和列索引進行切片選擇
print("用行和列索引進行切片選擇:")
print(df.loc[:, ['A', 'B']])
print("\n")

# 同時選擇行和列
print("同時選擇 '20230102' 行的 'A', 'B' 列:")
print(df.loc['20230102', ['A', 'B']])
print("\n")

# 使用 iloc 進行位置索引
print("使用 iloc 選擇第三行:")
print(df.iloc[3])
print("\n")

# 使用 iloc 進行位置索引的切片選擇
print("使用 iloc 選擇第三到第五行的第二到第三列:")
print(df.iloc[3:5, 1:3])
print("\n")

# 使用 iloc 選擇不連續的行和列
print("使用 iloc 選擇第一、三、五行的第二到第三列:")
print(df.iloc[[1, 3, 5], 1:3])
print("\n")

# 原始 DataFrame
print("原始 DataFrame:")
print(df)
print("\n")

# 使用條件選擇
print("選擇 'A' 列中大於 8 的行:")
print(df[df.A > 8])
````

`iloc` 和 `loc` 是 Pandas 中用於選擇和索引 DataFrame 的兩種方法，主要差異在於索引的方式。

1. **iloc (Integer-location based indexing):**
   - 使用整數位置索引（從 0 開始），類似於 NumPy 的索引方式。
   - 通過位置選擇單個元素、切片、布爾索引等。

   ```python
   df.iloc[行位置, 列位置]
   ```

2. **loc (Label-based indexing):**
   - 使用標籤索引，即行或列的標籤名稱。
   - 通過標籤選擇單個元素、切片、布爾索引等。

   ```python
   df.loc[行標籤, 列標籤]
   ```

簡而言之，`iloc` 使用整數位置索引，而 `loc` 使用標籤索引。以下是一些示例：

```python
# 使用 iloc 選擇第二行第三列的值
df.iloc[1, 2]

# 使用 loc 選擇 '20230102' 行的 'C' 列的值
df.loc['20230102', 'C']
```

需要注意的是，`iloc` 使用的是 Python 風格的索引，即半開區間 `[start, stop)`，而 `loc` 使用的是標籤索引，即閉區間 `[start, stop]`。

````python
import pandas as pd
import numpy as np

# 生成日期範圍
dates = pd.date_range('20230101', periods=6)

# 創建 DataFrame，行索引為日期，列索引為 'A', 'B', 'C', 'D'
df = pd.DataFrame(np.arange(24).reshape((6, 4)), index=dates, columns=['A', 'B', 'C', 'D'])
print("原始 DataFrame:")
print(df)
print("\n")

# 用 iloc 修改特定位置的值
df.iloc[2, 2] = 1111
print("用 iloc 修改特定位置的值:")
print(df)
print("\n")

# 用 loc 修改特定位置的值
df.loc['20230101', 'B'] = 2222
print("用 loc 修改特定位置的值:")
print(df)
print("\n")

# 用條件修改符合條件的值為 0
df[df.A > 4] = 0
print("將 'A' 列中大於 4 的值修改為 0:")
print(df)
print("\n")

# 將 'B' 列中對應 'A' 列中大於 4 的值修改為 0
df = pd.DataFrame(np.arange(24).reshape((6, 4)), index=dates, columns=['A', 'B', 'C', 'D'])
df.B[df.A > 4] = 0
print("將 'B' 列中對應 'A' 列中大於 4 的值修改為 0:")
print(df)
print("\n")

# 將 'A' 列中大於 4 的值修改為 0
df = pd.DataFrame(np.arange(24).reshape((6, 4)), index=dates, columns=['A', 'B', 'C', 'D'])
df.A[df.A > 4] = 0
print("將 'A' 列中大於 4 的值修改為 0:")
print(df)
print("\n")

# 新增一列 'F'，所有值為 NaN
df['F'] = np.nan
print("新增一列 'F'，所有值為 NaN:")
print(df)
print("\n")

# 新增一列 'E'，使用 pd.Series 進行賦值
df['E'] = pd.Series([1, 2, 3, 4, 5, 6], index=pd.date_range('20230101', periods=6))
print("新增一列 'E'，使用 pd.Series 進行賦值:")
print(df)
````

````python
import pandas as pd
import numpy as np

# 生成日期範圍
dates = pd.date_range('20230101', periods=6)

# 創建 DataFrame，行索引為日期，列索引為 'A', 'B', 'C', 'D'
df = pd.DataFrame(np.arange(24).reshape((6, 4)), index=dates, columns=['A', 'B', 'C', 'D'])

# 將某些值設置為 NaN
df.iloc[0, 1] = np.nan
df.iloc[1, 2] = np.nan

print("原始 DataFrame:")
print(df)
print("\n")

# 使用 dropna() 刪除包含 NaN 值的行
print("刪除包含 NaN 值的行（任意值為 NaN 就刪除整行）:")
print(df.dropna(axis=0, how='any'))
print("\n")

# 使用 dropna() 刪除所有值均為 NaN 的行
print("刪除所有值均為 NaN 的行（所有值都為 NaN 才刪除整行）:")
print(df.dropna(axis=0, how='all'))
print("\n")

# 使用 fillna() 將 NaN 值填充為特定值（這裡是填充為 0）
print("將 NaN 值填充為 0:")
print(df.fillna(value=0))
print("\n")

# 使用 isnull() 檢查每個元素是否為 NaN
print("檢查每個元素是否為 NaN:")
print(df.isnull())
print("\n")

# 使用 np.any() 判斷是否存在任何 NaN 值
print("判斷是否存在任何 NaN 值:")
print(np.any(df.isnull()) == True)

````
